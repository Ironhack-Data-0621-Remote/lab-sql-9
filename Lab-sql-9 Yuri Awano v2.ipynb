## Check the number of rentals for each customer for May.


```python
#QUERY FROM SQL
SELECT customer_id, count(customer_id) AS rent_num
FROM rental
WHERE rental_date REGEXP '^2005-05'
GROUP BY customer_id
ORDER BY rent_num DESC;
```

```python
query_3 = ''' SELECT customer_id, count(customer_id) AS rent_num
FROM rental
WHERE rental_date REGEXP '^2005-05'
GROUP BY customer_id
ORDER BY rent_num DESC '''
```

```python
data_count_may = pd.read_sql_query(query_3, engine)
data_count_may
```

## Check the number of rentals for each customer for June.
