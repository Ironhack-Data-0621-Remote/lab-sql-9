## Check the number of rentals for each customer for May.


```python
query_5 = ''' SELECT customer_id, count(customer_id) AS rent_num
FROM rental
WHERE rental_date REGEXP '^2005-05'
GROUP BY customer_id
ORDER BY rent_num DESC '''

data_count_may = pd.read_sql_query(query_5, engine)
data_count_may
```

## Check the number of rentals for each customer for June.
 

```python
query_6 = ''' SELECT customer_id, count(customer_id) AS rent_num
FROM rental
WHERE rental_date REGEXP '^2005-06'
GROUP BY customer_id
ORDER BY rent_num DESC '''

data_count_jun = pd.read_sql_query(query_6, engine)
data_count_jun
```
