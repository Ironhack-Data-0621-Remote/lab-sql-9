## Check the number of rentals for each customer for May.


```python
n_customer_rentals_may = '''
    SELECT customer_id, count(rental_id) as n_rental from rentals_may 
    GROUP BY customer_id
    ORDER BY n_rental desc
    '''

n_customer_rentals_may = pd.read_sql_query(n_customer_rentals_may, engine)

n_customer_rentals_may.head()
```

## Check the number of rentals for each customer for June.
